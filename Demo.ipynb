# Pillow

```python
import PIL
from PIL import Image
from PIL import ImageFilter
from IPython.display import display
```

```python
orginal_lion = Image.open('read_only/lion.jpg')
print(orginal_lion.format,orginal_lion.size,orginal_lion.mode)
display(orginal_lion)
```

# Thumbnail

```python
size = (128,128)
saved = "read_only/tumbnailLion.jpg"
try:
    orginal_image = Image.open('read_only/lion.jpg')
except:
    print("unable to load image")
    
orginal_image.thumbnail(size)
orginal_image.save(saved)
display(orginal_image)
```

# Gray Scale

```python
grayScale_lion = orginal_lion.convert("L")
grayScale_lion.save("read_only/grayScale_lion.jpg")
display(grayScale_lion)
```

 # Blure Filter

```python
blured_lion = orginal_lion.filter(ImageFilter.BLUR) 
blured_lion.save('read_only/blured_lion.jpg')
display(blured_lion)
```

```python
blured_gray_lion = grayScale_lion.filter(ImageFilter.BLUR) 
blured_gray_lion.save('read_only/blured_gray_lion.jpg')
display(blured_gray_lion)
```

# Contour Filter
